# 实训1 创建数组并进行运算

```python
import numpy as np
```

```python
 # 创建一个数值范围为0～1，间隔为0.01的数组
arr1 = np.arange(0, 1, 0.01)
print('创建的数组为：\n', arr1)
```

```python
# 创建100个服从正态分布的随机数数组
arr2 = np.random.randn(100)
print('创建的服从正态分布的随机数数组为：\n', arr2)
```

```python
# 数组的四则运算
print('数组相加的结果为：\n', arr1+arr2)
```

```python
# 数组相减
print('数组的相减的结果为：\n', arr1-arr2)
```

```python
# 数据相乘
print('数组相乘的结果为：\n', arr1*arr2)
```

```python
# 数据相除
print('数组相除的结果为：\n', arr1/arr2)
```

```python
# 数组幂运算
print('数组幂运算的结果为：\n', arr1**arr2)
```

```python
# 数组排序
print('原始数组为：\n', arr1)
print('排序后的数组为：\n', np.sort(arr1))
print('去重后的数组为：\n', np.unique(arr1))
```

```python
arr1.shape = 10, 10
print('重新设置shape的数组为：', arr1)
```
